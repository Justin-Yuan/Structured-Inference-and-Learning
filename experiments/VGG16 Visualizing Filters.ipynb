# Visualizing CNN Filters (raw weights)

Convolutional Neural Networks can be great on image related tasks, but what exactly is happening in the network? Alternatively, if the network indeed learns to recognize pictures, what is it learning?     

It turns out that various filters are adapting themselves to specific features of an image, from low-level patterns such as edges and lines near the input layer to higher-level, complex geometries like eyes or even wrinkles from deeper layers. A network feed-forward process essentially extracts corresponding features of an image in a hierarchical manner. Think of it as transforming an image from its pixel space to the feature space defined by te CNN. And with the new representation in the feature space, the image is ready for further tasks(classification, segmentation, etc).   

## Visualizing filters (mostly for the 1st Conv layer)

The following demonstrates how to visualize and understand what each filter/layer learns by layer raw weights. The raw weights may indicate portion of the image where the CNN pays much attention to, and hence could be seen as the pattern it learns. This is the most intuitive approach yet very limited in practice since modern CNN architectures tend towards smaller filters and deeper layers. Besides, filters from the first Convolutionala layer (the one right after the input layer) could be very informative about the input iamges, but filters from deeper layers are hard to interpret with respect to the original image since they are based on the transformed otuput from previous convolutional layer.  

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt 
%matplotlib inline

from keras.applications.vgg16 import preprocess_input 
from keras.preprocessing import image
```

## VGG16

VGG16 is a deep Convolution Neural Network that won the 2014 ImageNet ILSVRC competition. refer to the original paper (https://arxiv.org/pdf/1409.1556.pdf) or the Visual Geometry Group home page (http://www.robots.ox.ac.uk/~vgg/research/very_deep/) for more information.   

We are using a pretrained VGG16 model from Keras for its well defined architecture, relatively good performance and easy extensibility.

```python
from keras.applications.vgg16 import VGG16

model = VGG16(weights='imagenet')
```

create a VGG16 model with pretrained weights on ImageNet, include fully connected layers at the top. 

```python
# create a dictionary to easily refer to different network layers by name.
layer_dict = dict([(layer.name, layer) for layer in model.layers])

model.summary()
```

## Retrieve layer weights

To get the layer parameters, Keras provides the <a href="https://keras.io/layers/about-keras-layers/"> layer.get_weights() </a> function which returns both the filter weights and biases in a list.

```python
def get_layer_weights(layer_dict, layer):
    return layer_dict[layer].get_weights()

block1_conv1_weights = get_layer_weights(layer_dict, "block1_conv1")[0]
block1_conv2_weights = get_layer_weights(layer_dict, "block1_conv2")[0]

block4_conv1_weights = get_layer_weights(layer_dict, "block4_conv1")[0]
block4_conv2_weights = get_layer_weights(layer_dict, "block4_conv2")[0]
```

Tip: validate the layer weights by checking their dimensions with the model summary

```python
print(block1_conv1_weights.shape)
print(block1_conv2_weights.shape)
print(block4_conv1_weights.shape)
print(block4_conv2_weights.shape)
```

## Visualize filters weights 

```python
def visualize_filters(filter_weights, depth_index):
    size = filter_weights.shape[3]
    plt.figure(figsize=(20, 100))
    for i in range(size):
        plt.subplot(int(size/5)+1, 5, i+1)
        plt.imshow(filter_weights[:,:,depth_index,i])
    plt.show()
```

```python
# filters from the first convolutional layers 
visualize_filters(block1_conv1_weights, 2)
```

```python
# filters from a deeper convolutional layers 
visualize_filters(block1_conv2_weights, 1)
```

As you can see, each filter is typically a 3x3 grid image, which is not very interpretive as what feature(s) they learnt. Some very descriptive result was achieved by the AlexNet paper (https://papers.nips.cc/paper/4824-imagenet-classification-with-deep-convolutional-neural-networks.pdf), with 11x11 sized filters where you can see strip patterns. But in general, directly visualizing filter weights has limited use.
